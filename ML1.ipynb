```python
import pandas as pd
data = {"Item": ["TV", "TV", "TV", "AC"],"Company": ["LG", "VIDEOCON", "LG", "SONY"],"Rupees": [12000, 10000, 15000, 14000],"USD": [700, 650, 800, 750]}
df = pd.DataFrame(data)
print(df)
```

```python
new_data = pd.DataFrame([
    {"Item": "TV", "Company": "SONY", "Rupees": 13000, "USD": 720},
    {"Item": "AC", "Company": "LG", "Rupees": 16000, "USD": 850}
])
df = pd.concat([df, new_data], ignore_index=True)
print(df)
```

```python
lg_tv_max = df[(df['Item'] == 'TV') & (df['Company'] == 'LG')]['Rupees'].max()
print("Maximum price of LG TV:", lg_tv_max)
```

```python
total_rupees = df['Rupees'].sum()
print("Total sum of all products in Rupees:", total_rupees)
```

```python
sony_median_usd = df[df['Company'] == 'SONY']['USD'].median()
print("Median USD for SONY products:", sony_median_usd)
```

```python
sorted_df = df.sort_values(by='Rupees')
print(sorted_df)
```

```python
sorted_df = df.sort_values(by='Rupees')
print(sorted_df)
```

```python
import numpy as np
data = {"Name": ["Aparna", "Pankaj", "Ram", "Ramesh", "Naveen", "Krrishnav", "Bhawna"],"Degree": ["MBA", "BCA", "M.Tech", "MBA", np.nan, "BCA", "MBA"],"Score": [90.0, np.nan, 80.0, 98.0, 97.0, 78.0, 89.0]}
df2 = pd.DataFrame(data)
print(df2)
```

```python
print(df2.groupby("Degree")["Score"].max())
```

```python
df_filled = df2["Score"].fillna(76)
print(df_filled)
```

```python
df2.set_index("Name", inplace=True)
print(df2)
```

```python
df2.reset_index(inplace=True)
avg_scores = df2.groupby(["Name", "Degree"])["Score"].mean()
print(avg_scores)
```

```python
mba_count = df2[df2["Degree"] == "MBA"].shape[0]
print("Number of MBA students:", mba_count)
```

```python
bca_mode = df2[df2["Degree"] == "BCA"]["Score"].mode()
print("Mode of BCA scores:", bca_mode.values)
```

```python
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                'acceleration', 'model_year', 'origin', 'car_name']
autodf = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', names=column_names, delim_whitespace=True, na_values='?')
```

```python
print(autodf.head(10))
```

```python
print(autodf.isnull().sum())
```

```python
autodf_ffill = autodf.fillna(method='ffill')
```

```python
autodf_dropna = autodf.dropna()
```

```python
max_mpg_row = autodf.loc[autodf['mpg'].idxmax()]
print(max_mpg_row)
```

```python
avg_displacement = autodf.groupby('cylinders')['displacement'].mean()
print(avg_displacement)
```

```python
avg_cylinders = autodf['cylinders'].mean()
print("Average number of cylinders:", avg_cylinders)
```

```python
average_weight = autodf['weight'].mean()
count_heavy_cars = (autodf['weight'] > average_weight).sum()
print("Number of cars with weight greater than average:", count_heavy_cars)
```
